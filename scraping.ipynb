```python
import requests
from bs4 import BeautifulSoup
```

```python
#fetching data

page_count=1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore...")
        break

    with open(f"scraped_data/quotes{page_count}.html", "w", encoding="utf-8") as f:
              f.write(res.text)
              print(f"downloaded data from page{page_count}")

    page_count =page_count+1

              

    
    
```

```python
#open save files

with open("scraped_data/quotes1.html", "r", encoding="utf-8") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
#scraping one page
all_quotes = soup.select("div.quote")
life_quotes=[]

for q in all_quotes:
    all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author =  q.select_one("small.author").get_text()
        life_quotes.append([text , author])

print(life_quotes)
        
       
```

```python
#scraping all pages
life_quotes = []

for i in range(1, 11):
    with open(f"scraped_data/quotes{i}.html", "r", encoding="utf-8") as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, "lxml")
    all_quotes = soup.select("div.quote")

    for q in all_quotes:
        all_tags = [tag.get_text() for tag in q.select(".tag")]

        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

for list in life_quotes:
    print(list)
```

```python
#scraping all pages without writing 
page_count = 1
life_quotes = []

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore...")
        break

    for q in quotes:
        all_tags = [tag.get_text() for tag in q.select(".tag")]

        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

    print(f"processed page {page_count}")
    page_count += 1

print(life_quotes)
```

```python
#save data to csv
import csv
with open("life_quotesAll.csv" , "w", newline="", encoding="utf-8") as f:
    writer=csv.writer(f)

    #column
    writer.writerow(["Quote", "Author"])

    # data in table
    writer.writerows(life_quotes)

print("data saved to CSV successfully")
```
